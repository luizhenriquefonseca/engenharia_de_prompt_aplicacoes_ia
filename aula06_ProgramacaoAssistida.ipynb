Soma de Lista

```python
# Define uma lista de números inteiros
numeros = [1, 3, 6, 7, 8]

# Inicializa uma variável para armazenar a soma dos números
soma = 0

# Itera sobre cada número na lista 'numeros'
for num in numeros:
  # Adiciona o número atual à variável 'soma'
  soma += num

# Imprime o valor final da soma de todos os números
print(soma)
```
