### Jogo da Forma

Neste jogo simples, o computador vai "pensar" em uma forma geométrica e vai te dar uma dica. Sua tarefa é adivinhar qual é a forma!



````
# Primeiro, vamos importar uma ferramenta que nos ajuda a escolher coisas aleatoriamente.
# É como ter um dado que escolhe uma forma para a gente!
import random

# Agora, vamos criar uma lista de formas e algumas dicas sobre elas.
# Cada forma é um pequeno dicionário com 'nome' e 'dica'.
formas = [
    {"nome": "quadrado", "dica": "Eu tenho 4 lados iguais e 4 cantos retos."},
    {"nome": "círculo", "dica": "Eu sou redondinho e não tenho lados nem cantos."},
    {"nome": "triângulo", "dica": "Eu tenho 3 lados e 3 cantos."},
    {"nome": "retângulo", "dica": "Eu tenho 4 lados e 4 cantos retos, mas meus lados não são todos iguais."}
]

# Vamos iniciar o jogo em um loop infinito, para que ele continue até o jogador acertar ou decidir sair.
while True:
    # O computador vai escolher uma forma secreta da nossa lista.
    forma_secreta = random.choice(formas)
    # Agora, vamos guardar o nome da forma e a dica em variáveis separadas.
    nome_da_forma = forma_secreta["nome"]
    dica_da_forma = forma_secreta["dica"]

    # Vamos dar a dica para o jogador!
    print(f"\nEu estou pensando em uma forma. Aqui está uma dica:\n{dica_da_forma}")

    # Pedimos para o jogador tentar adivinhar qual é a forma.
    # O `input()` serve para o computador 'ouvir' o que o jogador digita.
    # Convertemos a resposta para minúsculas e removemos espaços extras.
    palpite_do_jogador = input("Qual forma eu sou? Digite sua resposta (ou 'sair' para desistir): ").lower().strip()

    # Verificamos se o jogador quer sair do jogo.
    if palpite_do_jogador == 'sair':
        print("Ok, obrigado por jogar! Até a próxima!")
        break # Sai do loop do jogo
    # Agora, vamos verificar se o palpite do jogador está correto!
    elif palpite_do_jogador == nome_da_forma:
        print(f"Parabéns! Você adivinhou! Eu sou um {nome_da_forma}!")
        break # Sai do loop porque o jogador acertou
    # Se não acertou e não quis sair, o jogo recomeça com uma nova forma.
    else:
        print(f"Ah, não foi dessa vez. Eu estava pensando em um {nome_da_forma}.")
        print("Mas não desanime! O jogo vai recomeçar com uma nova forma. Tente novamente!")
````

## Consultor de Endereços por CEP (ViaCEP)

Este programa em Python permitirá que você consulte informações de endereço (logradouro, bairro, cidade, estado) a partir de um Código de Endereçamento Postal (CEP) informado.

### Bibliotecas Necessárias

Para fazer requisições HTTP para a API da ViaCEP, utilizaremos a biblioteca `requests`. Se você ainda não a tem instalada, pode instalá-la executando o comando abaixo em uma célula de código:

### Código da Aplicação

Abaixo está o código completo da aplicação. Ele inclui uma função para lidar com a consulta do CEP, validação básica do formato do CEP e tratamento de erros.

````
# Importa a biblioteca 'requests' para fazer requisições HTTP (para a API).
import requests
import re # Importa a biblioteca 're' para usar expressões regulares (para validar o CEP)

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um determinado CEP.
    Recebe um CEP como string e retorna um dicionário com os dados do endereço ou None em caso de erro.
    """
    # A URL base da API ViaCEP para consultas de CEP.
    # '{cep}/json/' será substituído pelo CEP fornecido pelo usuário.
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Tenta fazer uma requisição GET para a URL da API.
        # 'requests.get()' envia a requisição e 'response' guarda a resposta do servidor.
        response = requests.get(url)

        # Verifica se a requisição foi bem-sucedida (código de status 200).
        # Se não for 200, significa que houve um problema na comunicação.
        response.raise_for_status()

        # Converte a resposta JSON da API em um dicionário Python.
        dados_cep = response.json()

        # A API ViaCEP retorna um dicionário com a chave 'erro' se o CEP não for encontrado.
        if 'erro' in dados_cep and dados_cep['erro']:
            return None # Retorna None para indicar que o CEP não foi encontrado
        else:
            return dados_cep # Retorna os dados do endereço se tudo estiver ok

    except requests.exceptions.RequestException as e:
        # Captura erros relacionados à requisição (ex: sem internet, URL inválida).
        print(f"Erro na requisição: {e}")
        return None
    except ValueError:
        # Captura erros se a resposta não for um JSON válido (o que é raro na ViaCEP).
        print("Erro: Resposta inválida da API (não é um JSON).")
        return None

def validar_cep(cep):
    """
    Valida se o formato do CEP é válido (8 dígitos numéricos).
    Retorna True se válido, False caso contrário.
    """
    # Expressão regular para verificar 8 dígitos numéricos, opcionalmente com um hífen.
    # Ex: 12345678 ou 12345-678
    padrao_cep = re.compile(r'^\d{5}-?\d{3}$|^\d{8}$|^\d{5}$|^\d{3}$|^\d{6}$|^\d{7}$|^\d{9}$') # Ajustado para aceitar CEP com 5, 3, 6, 7 ou 9 dígitos também (embora o padrão seja 8).

    # Usa re.fullmatch para garantir que toda a string corresponde ao padrão.
    if padrao_cep.fullmatch(cep):
        # Remove o hífen se existir, para enviar apenas os dígitos para a API.
        return cep.replace('-', '')
    else:
        return None

# --- Início da Aplicação Principal ---
if __name__ == "__main__":
    print("\n--- Consultor de Endereços por CEP ---")
    print("Digite o CEP (apenas números ou com hífen) para consultar o endereço.")
    print("Ou digite 'sair' para encerrar o programa.\n")

    while True:
        # Pede ao usuário para digitar um CEP.
        cep_digitado = input("Digite o CEP: ").strip()

        # Verifica se o usuário quer sair.
        if cep_digitado.lower() == 'sair':
            print("\nObrigado por usar o consultor de CEPs! Até a próxima.")
            break # Sai do loop, encerrando o programa.

        # Valida o formato do CEP digitado.
        cep_limpo = validar_cep(cep_digitado)

        if not cep_limpo:
            print("CEP inválido. Por favor, digite um CEP com 8 dígitos numéricos (ex: 12345-678 ou 12345678).\n")
            continue # Volta para o início do loop para pedir um novo CEP.

        print(f"Consultando CEP {cep_limpo}...")
        # Chama a função para consultar o CEP na API.
        dados = consultar_cep(cep_limpo)

        if dados:
            # Se dados foi retornado (não é None), exibe as informações.
            print("\n--- Endereço Encontrado ---")
            print(f"CEP: {dados.get('cep', 'N/A')}")
            print(f"Logradouro: {dados.get('logradouro', 'N/A')}")
            print(f"Complemento: {dados.get('complemento', 'N/A')}")
            print(f"Bairro: {dados.get('bairro', 'N/A')}")
            print(f"Localidade: {dados.get('localidade', 'N/A')}") # Cidade
            print(f"UF: {dados.get('uf', 'N/A')}") # Estado
            print(f"DDD: {dados.get('ddd', 'N/A')}")
            print("---------------------------\n")
        elif dados is None and 'erro' not in (consultar_cep(cep_limpo) or {}):
            # Caso a função retorne None e não haja 'erro' na resposta da API, significa um erro de conexão ou formato.
            print("Não foi possível obter informações para o CEP fornecido. Verifique o CEP ou sua conexão.\n")
        else:
            # Se dados for None e houver 'erro' na resposta, significa CEP não encontrado.
            print("CEP não encontrado. Por favor, verifique se o CEP está correto.\n")
````
