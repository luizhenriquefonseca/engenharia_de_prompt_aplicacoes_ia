## Missão 01: Automação de Arquivos

### O Problema (Briefing)

Criar um programa que organize arquivos bagunçados em pastas automaticamente, baseando-se apenas na extensão do arquivo (ex: .pdf, .jpg).

```
import os
import shutil

# --- Configurações ---
# ATENÇÃO: Crie uma pasta de teste e coloque alguns arquivos nela para testar o código.
# Altere 'pasta_baguncada' para o caminho da sua pasta de teste.
# Exemplo: source_directory = '/content/minha_pasta_de_teste'
source_directory = '/content/minha_pasta_de_teste' # Altere para o caminho da pasta que você quer organizar

# Certifica-se de que a pasta de origem existe
if not os.path.exists(source_directory):
    print(f"A pasta de origem '{source_directory}' não existe. Por favor, crie-a e adicione alguns arquivos para testar.")
else:
    print(f"Organizando arquivos em: {source_directory}")

    # Itera sobre todos os arquivos na pasta de origem
    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Verifica se é um arquivo (e não um subdiretório)
        if os.path.isfile(file_path):
            # Obtém a extensão do arquivo (ex: '.pdf', '.jpg')
            file_extension = os.path.splitext(filename)[1].lower()

            # Se não houver extensão (ex: 'meu_arquivo_sem_extensao'), use 'sem_extensao'
            if not file_extension:
                destination_folder = os.path.join(source_directory, 'sem_extensao')
            else:
                # Cria o nome da pasta de destino (ex: '.pdf' -> 'pdfs')
                # Remove o ponto e adiciona um 's' para o nome da pasta
                destination_folder_name = file_extension[1:] + 's'
                destination_folder = os.path.join(source_directory, destination_folder_name)

            # Cria a pasta de destino se ela não existir
            os.makedirs(destination_folder, exist_ok=True)

            # Move o arquivo para a pasta de destino
            shutil.move(file_path, os.path.join(destination_folder, filename))
            print(f"Movido '{filename}' para '{destination_folder_name}'")

    print("Organização de arquivos concluída!")
```

# Medidor de temperatura

Este programa simples verifica a temperatura de um sistema e notifica o usuário se houver risco de superaquecimento, com base em um limite de 30 graus Celsius.

```
# Solicita ao usuário para inserir a temperatura atual do sistema
temperatura_atual = int(input("Por favor, insira a temperatura atual do sistema em Celsius: "))

# Verifica se a temperatura atual é maior que 30 graus Celsius
if temperatura_atual > 30:
    # Se a temperatura for maior que 30, exibe um alerta de superaquecimento
    print("ALERTA: Superaquecimento! Temperatura atual: " + str(temperatura_atual) + "°C")
else:
    # Se a temperatura não for maior que 30, exibe uma mensagem de temperatura normal
    print("Temperatura do sistema normal: " + str(temperatura_atual) + "°C")
```

##  Missão 03: Sistema de Notificações

### O Problema (Briefing)

Criar um programa autônomo que dispare alertas (mensagens de alerta no console ou e-mail simulado) unicamente quando uma condição crítica for atingida.

```
import time
import random

# Define o limite crítico de temperatura
TEMPERATURA_CRITICA = 30

# Define o número máximo de verificações antes de parar
MAX_CHECKS = 5
check_count = 0

print("Iniciando monitoramento autônomo do sistema... (Para parar, interrompa o kernel)")

while check_count < MAX_CHECKS:
    # Simula a leitura de temperatura do sistema (em um sistema real, isso viria de um sensor)
    temperatura_simulada = random.randint(20, 45) # Gera temperaturas entre 20 e 45 graus Celsius

    print(f"[{time.strftime('%H:%M:%S')}] Verificando... Temperatura atual: {temperatura_simulada}°C")

    # Verifica se a temperatura atingiu a condição crítica
    if temperatura_simulada > TEMPERATURA_CRITICA:
        mensagem_alerta = f"ALERTA CRÍTICO: Superaquecimento detectado! Temperatura: {temperatura_simulada}°C."
        print(mensagem_alerta)
        # Simula o envio de um e-mail de alerta
        print("Simulando envio de e-mail de alerta para 'admin@empresa.com' com a mensagem acima.")
    else:
        print("Temperatura do sistema normal.")

    # Incrementa o contador de verificações
    check_count += 1

    # Aguarda alguns segundos antes da próxima verificação para simular um intervalo
    time.sleep(5) # Verifica a cada 5 segundos

print("Monitoramento autônomo encerrado após atingir o número máximo de verificações.")
```
